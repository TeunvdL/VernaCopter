```python
import numpy as np
from make_shapes import shapes
```

```python
import matplotlib.pyplot as plt

center = [0, 0, 0]
length = 32 * 2
width = 50 * 2
height = 100 * 2

fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(projection='3d') 
ax.set_aspect("equal")

# set axes limits
axis_size = 200
xcenter, ycenter, zcenter = 50, 50, 0

ax.set_xlim3d([-axis_size/2 + xcenter, axis_size/2 + xcenter])
ax.set_ylim3d([-axis_size/2 + ycenter, axis_size/2 + ycenter])
ax.set_zlim3d([-axis_size/2 + zcenter, axis_size/2 + zcenter])

X, Y, Z = shapes.make_cuboid(center, (length, width, height))
ax.plot_surface(X, Y, Z, color='b', rstride=1, cstride=1, alpha=0.2, linewidth=1., edgecolor='k')

# plot a sphere
center = [100, 100, 0]
X, Y, Z = shapes.make_sphere(center, 50)
ax.plot_surface(X,Y,Z, color='b', alpha=0.3)

plt.show()
```
